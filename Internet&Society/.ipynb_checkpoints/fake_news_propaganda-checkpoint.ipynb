```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Data Collection - 2020 US Elections (3).xlsx')
x = df["Times shared on Facebook(killo)"]
y = df["Times retweeted on Twitter(kilo)"]
c = df["Truth of the news (1 for true, -1 for fake, lie by omission 0)"]
side = df["Republicans Favor (R) or Democrats Favor (D)"]
```

```python
ones = np.where(c == 1)[0]
not_ones = np.where(c != 1)[0]
plt.figure(figsize=(10, 6))
plt.title("True and fake news about 2020 U.S. elections shared on social media from May 2nd 2020 until November 3rd 2020")
plt.plot(x[ones], y[ones], 'og')
plt.plot(x[not_ones], y[not_ones], 'or')
# ax = plt.gca()
# lims = [
#     np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
#     np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
# ]
# plt.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
plt.savefig('TruthOrNot.png')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Interactions on Facebook [log scale] (in Kilo)")
plt.ylabel("Interactions on Twitter [log scale] (in Kilo)")
_ = plt.legend(["true news", "fake news"])
```

```python
democrats = np.where(side == 'D')[0]
republicans = np.where(side == 'R')[0]
plt.figure(figsize=(11, 8))
plt.plot(x[republicans], y[republicans], 'or')
plt.plot(x[democrats], y[democrats], 'ob')
plt.savefig('DemocratsvRepublicans.png')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Interactions on Facebook [log scale] (in Kilo)")
plt.ylabel("Interactions on Twitter [log scale] (in Kilo)")
_ = plt.legend(["Republican news", "Democrats news"])
```

```python
democratsTrue = np.where((side == 'D') & (c == 1))[0]
democratsFalse = np.where((side == 'D') & (c != 1))[0]
plt.figure(figsize=(6, 6))
plt.title("True and fake news shared by Democrats party\nbetween May 2nd and November 3rd 2020")
plt.plot(x[democratsTrue], y[democratsTrue], '*b')
plt.plot(x[democratsFalse], y[democratsFalse], '+g')
plt.savefig('DemocratsAnalysis.png')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Interactions on Facebook [log scale] (in Kilo)")
plt.ylabel("Interactions on Twitter [log scale] (in Kilo)")
_ = plt.legend(["true news", "fake news"])
```

```python
republicansTrue = np.where((side == 'R') & (c == 1))[0]
republicansFalse = np.where((side == 'R') & (c != 1))[0]
plt.figure(figsize=(6, 6))
plt.title("True and fake news shared by Republicans party\nbetween May 2nd and November 3rd 2020")
plt.plot(x[republicansTrue], y[republicansTrue], '*r')
plt.plot(x[republicansFalse], y[republicansFalse], '+g')
plt.savefig('republicansAnalysis.png')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Interactions on Facebook [log scale] (in Kilo)")
plt.ylabel("Interactions on Twitter [log scale] (in Kilo)")
_ = plt.legend(["true news", "fake news"])
```
